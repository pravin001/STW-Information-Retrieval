```
# TF-IDF and Cosine Similarity Calculation
import math

# Term Frequencies
doc1 = {'student': 1, 'room': 4, 'coventry': 7, 'hub': 0}
doc2 = {'student': 2, 'room': 1, 'coventry': 1, 'hub': 2}
doc0 = {'student': 1, 'room': 1, 'hub': 1, 'coventry': 0}

# IDF values based on document frequency
idf = {
    'student': math.log10(5/4),
    'room': math.log10(5/4),
    'coventry': math.log10(5/4),
    'hub': math.log10(5/1)
}

# Compute TF-IDF vector
def compute_tf_idf(doc):
    return [doc.get(term, 0) * idf[term] for term in ['student', 'room', 'hub', 'coventry']]

# Dot product of two vectors
def dot(v1, v2):
    return sum(x * y for x, y in zip(v1, v2))

# Magnitude of a vector
def magnitude(v):
    return math.sqrt(sum(x ** 2 for x in v))

# Cosine similarity
def cosine_sim(v1, v2):
    return dot(v1, v2) / (magnitude(v1) * magnitude(v2))

# Compute TF-IDF vectors
v0 = compute_tf_idf(doc0)
v1 = compute_tf_idf(doc1)
v2 = compute_tf_idf(doc2)

# Cosine similarities
cos1 = cosine_sim(v0, v1)
cos2 = cosine_sim(v0, v2)

print('TF-IDF Vector for Doc0:', v0)
print('TF-IDF Vector for Doc1:', v1)
print('TF-IDF Vector for Doc2:', v2)
print('Cosine Similarity (Doc0, Doc1):', round(cos1, 4))
print('Cosine Similarity (Doc0, Doc2):', round(cos2, 4))
```
